```python
# Matplotlib usage - the basics

import matplotlib.pyplot as plt
import numpy as np

# let jupyter show the plots in the webpage itself
%matplotlib inline
```

```python
x = np.linspace(-6, 6, 180)
y = np.sin(x)
```

```python
# Functional method
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine wave')
plt.grid(True)
```

```python
# subplots - in the 2x1 layout

# select first plot in the 2x1 array
plt.subplot(2,1,1)
plt.plot(x, y, 'b')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine wave')
plt.grid(True)

z = np.cos(x)
# now select second plot in the 2x1 array
plt.subplot(2,1,2)
plt.plot(x, z, 'r')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Cosine wave')
plt.grid(True)
```

```python
# Object-oriented method
# Manually scale the plot dimensions

figure = plt.figure()
sin_axes = figure.add_axes([0.0, 0.6, 0.9, 0.4]);
sin_axes.plot(x,y)
sin_axes.set_xlabel('x')
sin_axes.set_ylabel('sin(x)')
sin_axes.set_title('Sine wave')
sin_axes.grid(True)

cos_axes = figure.add_axes([0.0, 0.0, 0.9, 0.4]);
cos_axes.plot(x,z)
cos_axes.set_xlabel('x')
cos_axes.set_ylabel('cos(x)')
cos_axes.set_title('Cosine wave')
cos_axes.grid(True)
```

```python
# Object oriented method - better way
figure, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,9))

# relaxes the layout for better appearance
plt.tight_layout()

# sin(x)
axes[0][0].plot(x,y,'b--')
axes[0][0].set_xlabel('x')
axes[0][0].set_ylabel('sin(x)')
axes[0][0].set_title('Sine wave')
axes[0][0].grid(True)

# cos(x)
axes[1][0].plot(x,z, 'r--')
axes[1][0].set_xlabel('x')
axes[1][0].set_ylabel('cos(x)')
axes[1][0].set_title('Cosine wave')
axes[1][0].grid(True)

# sin(x) * sin(x)
y2 = y ** 2
axes[0][1].plot(x,y2,'b--')
axes[0][1].set_xlabel('x')
axes[0][1].set_ylabel('sin(x) * sin(x)')
axes[0][1].set_title('Sin^2(x) square')
axes[0][1].grid(True)

# cos(x) * cos(x)
z2 = z ** 2
axes[1][1].plot(x,y2,'r--')
axes[1][1].set_xlabel('x')
axes[1][1].set_ylabel('cos(x) * cos(x)')
axes[1][1].set_title('Cos^2(x) square')
axes[1][1].grid(True)
```

```python
# sin^2(x) + cos^2(x) = 1
# Object oriented method - better way
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))

plt.tight_layout()

# sin^2(x) + cos^2(x)
axes[0].plot(x,(y2+z2),'b--')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin^2(x) + cos^2(x)')
axes[0].set_title('Sum of Sin^2(x) and Cos^2(x)')
axes[0].grid(True)

# sin^2(x) - cos^2(x)
axes[1].plot(x,(y2-z2), 'r--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('sin^2(x) - cos^2(x)')
axes[1].set_title('Difference of Sin^2(x) and Cos^2(x)')
axes[1].grid(True)
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax = fig.add_axes([0,0,1,1])

ax.set_xlim([-2,2])
ax.set_ylim([-2,2])

# concentric circle
# this is drawn using points - there are other better ways to do it
x = np.linspace(-1,1,50);
y = np.sqrt(1 - x**2);
ax.plot(x, y, color='blue');
ax.plot(x, -y, color='blue');

x = np.linspace(-2,2,50);
y = np.sqrt(4 - x**2);
ax.plot(x, y, color='red');
ax.plot(x, -y, color='red');

# parabola and inverted parabola
x = np.linspace(-2,2,50);
ax.plot(x, x**2, color='blue', ls='--', lw=2)
ax.plot(x, -x**2, color='orange', ls='--', lw=2)

# y = x
ax.plot(x, x, color='black')

ax.grid(axis='both');

```
